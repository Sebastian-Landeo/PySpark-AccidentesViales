```python
import pdfplumber
import pandas as pd
import os

# Lista de tus archivos PDF por tipo
tipos = ["A", "B", "C", "D", "E", "F", "G"]
folder = "./"  # Cambia a donde tengas tus PDFs
data = []

for tipo in tipos:
    file_path = os.path.join(folder, f"Tipo {tipo}.pdf")

    print(f"Procesando Tipo {tipo}...")

    with pdfplumber.open(file_path) as pdf:
        for page in pdf.pages:
            table = page.extract_table()
            if not table:
                continue

            for row in table[1:]:  # Saltamos cabecera
                if len(row) < 5:
                    continue

                ubigeo = str(row[1]).strip().zfill(6)
                if ubigeo.isdigit():
                    data.append({"ubigeo": ubigeo, "clasificacion_muni": tipo})

# Crear DataFrame final
df = pd.DataFrame(data)

# Agregar manualmente los 21 registros faltantes con su clasificación correcta
faltantes = [
    {"ubigeo": "021306", "clasificacion_muni": "G"},
    {"ubigeo": "120108", "clasificacion_muni": "F"},
    {"ubigeo": "050406", "clasificacion_muni": "G"},
    {"ubigeo": "090718", "clasificacion_muni": "G"},
    {"ubigeo": "040310", "clasificacion_muni": "E"},
    {"ubigeo": "080805", "clasificacion_muni": "F"},
    {"ubigeo": "120125", "clasificacion_muni": "D"},
    {"ubigeo": "101101", "clasificacion_muni": "B"},
    {"ubigeo": "050206", "clasificacion_muni": "F"},
    {"ubigeo": "021511", "clasificacion_muni": "F"},
    {"ubigeo": "080906", "clasificacion_muni": "G"},
    {"ubigeo": "130108", "clasificacion_muni": "G"},
    {"ubigeo": "110105", "clasificacion_muni": "E"},
    {"ubigeo": "211202", "clasificacion_muni": "F"},
    {"ubigeo": "210807", "clasificacion_muni": "G"},
    {"ubigeo": "190205", "clasificacion_muni": "E"},
    {"ubigeo": "160604", "clasificacion_muni": "G"},
    {"ubigeo": "150203", "clasificacion_muni": "E"},
    {"ubigeo": "240301", "clasificacion_muni": "A"},
    {"ubigeo": "060611", "clasificacion_muni": "G"},
    {"ubigeo": "151016", "clasificacion_muni": "F"},
]

df = pd.concat([df, pd.DataFrame(faltantes)], ignore_index=True)

# Eliminar duplicados por si aparecen en varias páginas o en los manuales
df = df.drop_duplicates(subset=["ubigeo"])

# Mostrar primeras filas para verificar
print(df.head())

# Guardar a CSV
df.to_csv("clasificacion_munies_abcg.csv", index=False)
print("CSV generado exitosamente")
```
