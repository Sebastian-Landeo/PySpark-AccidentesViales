```python
import os
import pandas as pd

input_folder = "AccidentesRaw/"
output_folder = "Accidentes/"

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.endswith(".xlsx") or file.endswith(".xls"):
        excel_path = os.path.join(input_folder, file)
        print(f"üìÑ Leyendo Excel: {excel_path}")

        try:
            # leer sin header para buscarlo manualmente
            temp = pd.read_excel(excel_path, header=None)

            # encontrar fila donde est√° "C√ìDIGO SINIESTRO"
            header_row = temp[temp.apply(lambda row: row.astype(str).str.contains("C√ìDIGO SINIESTRO").any(), axis=1)].index[0]

            # leer nuevamente usando esa fila como encabezado
            df = pd.read_excel(excel_path, header=header_row)

            # guardar csv
            csv_name = file.rsplit(".", 1)[0] + ".csv"
            csv_path = os.path.join(output_folder, csv_name)

            df.to_csv(csv_path, index=False, encoding="utf-8")
            print(f"‚úÖ Guardado como CSV: {csv_path}")

        except Exception as e:
            print(f"‚ùå Error procesando {file}: {e}")
```

```python
import os
import chardet

input_folder = "Red_VialRaw/"
output_folder = "Red_Vial/"

os.makedirs(output_folder, exist_ok=True)

for file in os.listdir(input_folder):
    if file.endswith(".csv"):
        input_path = os.path.join(input_folder, file)
        output_path = os.path.join(output_folder, file)

        print(f"üìÑ Procesando: {input_path}")

        # Detectar encoding
        with open(input_path, "rb") as f:
            raw = f.read()
            detected = chardet.detect(raw)
            encoding = detected["encoding"]

        print(f"üîé Encoding detectado: {encoding}")

        try:
            # Leer usando el encoding detectado
            text = raw.decode(encoding)

            # Guardar en UTF-8
            with open(output_path, "w", encoding="utf-8", newline="") as f:
                f.write(text)

            print(f"‚úÖ Guardado en UTF-8: {output_path}")

        except Exception as e:
            print(f"‚ùå Error al convertir {file}: {e}")
```
