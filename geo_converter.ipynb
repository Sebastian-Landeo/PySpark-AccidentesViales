```python
import geopandas as gpd

gdf = gpd.read_file("Red_VialRaw/Geo/Red Vial Nacional 2024 MTC PROVIAS GEO GPS PERU Suyo Pomalia.shp")
print(gdf.head())
```

```python
# columnas útiles
cols = [
    "OBJECTID", "Id", "cCodRuta", "cNomRuta", "cIdTramo",
    "dkmInicio", "dkmFinal", "dLongitud", "dNroCarril", "dAncCalzad",
    "cTipRed", "cClasifica", "cAutopista", "cSentido",
    "cCodDepart", "cDepartame", "cCodProvin", "cRegion", "cTopografi",
    "cEstado",
    "dSuperfici", "dLogistico", "cOperacion", "dIMD",
    "cPeajes"
]

# filtrar solo columnas existentes
cols = [c for c in cols if c in gdf.columns]
df = gdf[cols]

# llenar valores nulos con string vacío
df[str_cols] = df[str_cols].fillna("").astype(str)

# reemplazar strings vacíos en la columna cPeajes por "Sin peaje"
def limpiar_peaje(x):
    if x in ["", None]:
        return "Sin peaje"
    elif x == "RN":
        return "Red Nacional"
    elif "Concesión" in str(x):  # con tilde
        return "Concesión"
    elif str(x).isdigit():
        return "Peaje"
    else:
        return str(x)  # cualquier otro valor inesperado

df['cPeajes'] = df['cPeajes'].apply(limpiar_peaje)



# convertir tipos
int_cols = ["OBJECTID", "Id", "cIdTramo", "dNroCarril", "cCodDepart", "cCodProvin"]
float_cols = ["dkmInicio", "dkmFinal", "dLongitud", "dAncCalzad", "dSuperfici", "dLogistico", "dIMD"]
str_cols = [c for c in df.columns if c not in int_cols + float_cols]

# aplicar conversiones
df[int_cols] = df[int_cols].fillna(0).astype(int)
df[float_cols] = df[float_cols].astype(float)
df[str_cols] = df[str_cols].fillna("").astype(str)

# exportar
df.to_csv("Red_Vial/RedVial_2024_clean.csv", index=False)
print("CSV generado con tipos asegurados.")
```
